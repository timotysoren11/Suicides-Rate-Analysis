```python
# Creating duplicate of dataset

df_data = data.copy()
df_data.head()
```

```python
# checking top 5 countries with highest suicide rate

suicide_rate_top5 = df_data.sort_values(by='suicides_no', ascending=False).head(5)
suicide_rate_top5
```

```python
df_data.dtypes
```

```python
# Displaying the number of suicides took place in some year

# Aggregate the data by year and sum the number of suicides
suicides_by_year = df_data.groupby('year')['suicides_no'].sum()                         
suicides_by_year
```

```python
# Find the year with the maximum number of suicides
year_with_most_suicides = suicides_by_year.idxmax()
year_with_most_suicides
```

```python
# Display the year with the most suicides and the corresponding count
suicides_by_year.max()
```

```python
# checking number of suicides according to age group
suicides_by_age = df_data.groupby('age')['suicides_no'].sum() 
suicides_by_age
```

```python
# Removing the duplicate countries name from the data set
suicide_by_country = df_data.drop_duplicates('country')
suicide_by_country
```

```python
# Removing the index number 
# Remove the default integer index
removeing_default_index = suicide_by_country.reset_index(drop=True)
removeing_default_index
```

```python
indexed_data = removeing_default_index.copy()
```

```python
indexed_data.head()
```

```python
# Top 5 countries with highest suicides case
indexed_data = indexed_data.sort_values(by='suicides_no', ascending=False)
indexed_data.head()
```

```python
# Countries with lowest suicide numbers
indexed_data = indexed_data.sort_values(by='suicides_no', ascending=True)
indexed_data.head()
```

# Creating new dataset from the existing one 

```python
# Selecting the desired columns
selected_columns = ['country','year','age','suicides_no','generation']

# Stroing it into new dataset
new_dataset = indexed_data[selected_columns]

# Saving the new dataset to a csv file
new_dataset.to_csv('suicides Rate.csv', index=False)
```

# Importing the newly created csv file

```python
df_data = pd.read_csv('suicides Rate.csv')
df_data
```

```python
df_data.head()
```

# Visualization

```python
# Creating lineplot - shows changes in suicide over time period
sns.lineplot(x='year', y='suicides_no', hue='generation', data=df_data)
plt.title('Suicide Rates Over Time')
plt.show()
```

```python
# Bar chart for comaprision of dataset on the basis of age
sns.barplot(x='age', y='suicides_no', data=df_data)
plt.title('Suicide Rates by Category')
plt.show()
```

```python
# box-plot - showing distribution of suicide numbers
sns.boxplot(x='generation', y='suicides_no', data=df_data)
plt.title('Distribution of Suicide Rates')
plt.show()
```
